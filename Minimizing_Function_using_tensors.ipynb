```python
import tensorflow as tf
import numpy as np
import time
```

```python
x = tf.Variable(tf.random.normal((1,)))
lr = tf.constant(0.001)
tolerance = 1.

yprev = x**2-4.
while True:
  print("x,yprev:",x.numpy(),yprev.numpy())
  with tf.GradientTape() as tape:
    y = x**2 - 4.
  
  dy_dx = tape.gradient(y,x)
  x.assign(x - lr*dy_dx)
  yNew = x**2 - 4.
  print("x,ynew:",x.numpy(),yNew.numpy())
  tol = 100*tf.abs((yNew - yprev)/yprev).numpy()[0]

  print("delta:",tol)
  if tol<=tolerance:
    print("Exiting loop")
    break
  else:
    yprev = yNew
    time.sleep(1)
```

```python
x = tf.Variable(tf.random.normal((1,)))
lr = tf.constant(0.1)

yprev = 7*(x**2)-(4*x)+10.
num_iter=1
while num_iter<=10:
  print("Iteration:",num_iter)
  print("x,yprev:",x.numpy(),yprev.numpy())
  with tf.GradientTape() as tape:
    y = 7*(x**2)-(4*x)+10
  
  dy_dx = tape.gradient(y,x)
  x.assign(x - lr*dy_dx)
  yNew = 7*(x**2)-(4*x)+10.
  print("x,ynew:",x.numpy(),yNew.numpy())
  yprev = yNew
  time.sleep(1)
  num_iter+=1
```
